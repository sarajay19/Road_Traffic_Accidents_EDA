# Tasks

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Traffic.csv')
```

## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
df.info()
```

```python
df.head()
```

```python
df.groupby('City')['TrafficIndexWeekAgo'].value_counts()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='TrafficIndexLive', kde=True)
plt.title('Distribution of Traffic')
plt.show()
```

```python
# plt.figure(figsize=(12, 6))
# sns.histplot(data=df, x='Datetime', kde=True)
# plt.title('Distribution of Datetime')
# plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='TrafficIndexLive', y='TravelTimeLive')
plt.title('Relationship between Traffic and Weather')
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='City', y='TrafficIndexLive')
plt.title('Traffic Index Live by City')
plt.xticks(rotation=90)
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
df['TrafficIndexLive'].value_counts().plot(kind='bar')
plt.title('Traffic Index Live')
plt.xlabel('Traffic')
plt.ylabel('Count')
plt.show()
```
